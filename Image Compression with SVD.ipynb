```python
import numpy as np
from numpy.linalg import svd
from numpy import array
from numpy import diag
from numpy import zeros
from matplotlib import image
from matplotlib import pyplot
from skimage import io, color
```

```python
X=io.imread('images\dog1.png')
X=color.rgb2gray(X)
pyplot.imshow(X,cmap='gray')
print(X.shape)
```

```python
U, s, VT=svd(X)  #SVD
print(U.shape)
print(s.shape)
print(VT.shape)
```

```python
Sigma=zeros(X.shape)  # X is our original mxn matrix
#And lets diagonalize s
s=diag(s)
s.shape
```

```python
Sigma[:s.shape[0],:s.shape[1]]=s
Sigma.shape
```

```python
X_rec=U@Sigma@VT
print("Reconstructed matrix:\n")
pyplot.imshow(X_rec,cmap='gray')
```

```python
range_=[1,3,10,30]
for i in range_:
    k=i
    print("value of k:",k)
    X_approximate=U@Sigma[:,:k]@VT[:k,:]
    pyplot.imshow(X_approximate,cmap='gray')
    pyplot.show()
```
